```python
import pandas as pd 
```

```python
file = pd.read_csv('D:\\datasets\\car_price\\car_price.csv')
df = file.copy()
df.sample(10)
```

```python
df.info()
```

```python
#seperating columns with nan data type
nan_cols = list()
for i in df.columns :
    if df[i].dtype == 'O' :
        nan_cols.append(i)
nan_cols
```

```python
#calculate the value count of df in nan columns
for i in nan_cols :
    print(i)
    print(len(list(set(df[i]))))
```

```python
label = df['price']
df = df.drop(['price'],axis=1)
num_df = df.drop(nan_cols,axis=1)
nan_df = df.drop(num_df.columns,axis=1)
num_df.sample(10)
```

```python
num_df['label'] = label
Corr = num_df.corr()
Corr['label'].sort_values(ascending=False)*100
```

```python
from matplotlib.pyplot import show ,scatter , xlabel , ylabel
for column in num_df.columns :
    if column == 'label' :
        continue
    ylabel('price')
    xlabel(column)
    scatter(num_df[column],num_df['label'])
    show()
```

```python
num_df = num_df.drop(['label'],axis=1)
```

```python
nan_df.sample(10)
```

```python
carname = nan_df['CarName']
fuelsystem = nan_df['fuelsystem']
nan_df = nan_df.drop(['CarName','fuelsystem'],axis=1)
nan_df.head()
```

```python
encoded_nan1 = pd.get_dummies(nan_df)
encoded_nan1.sample(10)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
encoded_carname = pd.DataFrame(encoder.fit_transform(carname),columns=['carname'])
encoded_fuelsystem = pd.DataFrame(encoder.fit_transform(carname),columns=['fuelsystem'])
finall_nan_df = pd.concat([encoded_nan1,encoded_carname,encoded_fuelsystem],axis=1)
finall_nan_df.sample(10)
```

```python
finall_nan_df.info()
```

```python
ready_df = pd.concat([num_df,finall_nan_df],axis=1)
ready_df.head()
```

```python
ready_df.info()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df = scaler.fit_transform(ready_df)
scaled_df[:3]
```

```python
from sklearn.model_selection import train_test_split
xtr , xts ,ytr ,yts = train_test_split(scaled_df,label,test_size=0.1,random_state=22)
if len(xtr) == len(ytr) :
    print('spliting done')
```

```python
from sklearn.linear_model import LinearRegression
alg = LinearRegression().fit(xtr,ytr)
alg.score(xtr,ytr)*100
```

```python
alg.score(xts,yts)*100
```

```python
alg.score(scaled_df,label)*100
```
